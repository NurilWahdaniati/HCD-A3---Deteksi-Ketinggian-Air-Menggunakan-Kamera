```python
import matplotlib.pyplot as plt  
import matplotlib.image as mping  
import cv2 

img = mping.imread ("sungaii.jpg") 
plt.imshow(img) 
```

```python
im = mping.imread ("sungaii.jpg")  
gray = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY) 
plt.imshow(gray) 
```

```python
gray = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR) 
plt.imshow(gray) 
plt.show()
```
